```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Lendo o dataframe preparado somente com os dados do Brasil

```python
# Lendo o dataframe
df_brasil = pd.read_excel('manipulacao_br_python.xlsx')

# Setando o .ipynb para exibir todas as colunas
pd.set_option('display.max_columns', None)

# Filtrando o dataframe para ignorar anos em que no h registro de regras fiscais em vigor
df_brasil = df_brasil[df_brasil['quantity_of_rules_in_place'] != 0]
df_brasil
```

# Trocando todos os valores '-' do df por '0'

```python
df_brasil.replace(to_replace = '-', value = 0, inplace = True)
df_brasil
```

# Calculando o IRF: 4 sub-ndices por ano

## Subndices:

* Supporting procedures
* Formal enforcement procedures
* Coverage
* Legal basis


## Anos com uma nica regra fiscal


```python
# Exibindo os anos correspondentes a essa condio

df_1rule = df_brasil[df_brasil['quantity_of_rules_in_place'] == 1]
df_1rule
```

```python
# Calculando o primeiro sub-ndice

multiyear= df_1rule['my_expc_agreggate'] + df_1rule['my_expc_ministries'] + df_1rule['my_expc_line-items']
# Normalizando a varivel:
multiyear = multiyear/3

independent_body = df_1rule['ind_body_sets_budg_assump'] + df_1rule['ind_body_monit_implement']
#Normalizando a variavel:
independent_body = independent_body/2

fiscal_responsability_law = df_1rule['fiscal_responsability_law']

# Construindo o sub-ndice j normalizado para [0,1]
supporting_procedures = (fiscal_responsability_law + independent_body + multiyear)/3 

# supporting_procedures = supporting_procedures.reset_index() isso transforma a series em um dataframe
supporting_procedures
```

```python
# Calculando o segundo sub-ndice normalizado (s tem 1 regra fiscal, ento no precisamos normalizar pra ficar de 0 a 1, ele j fica por si s)
enforc_procedures = (df_1rule['enforc_nat_DR'] + df_1rule['enforc_nat_ER'] + df_1rule['enforc_nat_BBR'] + df_1rule['enforc_nat_RR'])

```

```python
# Calculando o terceiro sub-ndice (s tem 1 regra fiscal, ento no precisamos normalizar pra ficar de 0 a 1, ele j fica por si s)
coverage = (df_1rule['cov_nat_DR']/2.5 + df_1rule['cov_nat_ER']/2.5 + df_1rule['cov_nat_BBR']/2.5 + df_1rule['cov_nat_RR']/2.5)
coverage
```

```python
# Calculadno o quarto sub-ndice (s tem 1 regra fiscal, ento no precisamos normalizar pra ficar de 0 a 1, ele j fica por si s)
legal_basis = (df_1rule['lb_nat_DR']/5 + df_1rule['lb_nat_ER']/5 + df_1rule['lb_nat_BBR']/5 + df_1rule['lb_nat_RR']/5)
legal_basis
```

```python
# Normalizando os quatro sub-ndices para o intervalo [0,5] e os somando
FRI = (supporting_procedures*5 + enforc_procedures*5 + coverage*5 + legal_basis*5)

# Normalizando o FRI para o intervlao [0,5]
FRI = FRI*5/20
FRI = pd.DataFrame(FRI)

# Transformando a series em um dataframe para colocarmos os outros resultados depois
FRI.set_index(df_1rule['Year'], inplace = True)
FRI.rename({0: 'FRI'}, axis = 1, inplace = True)
FRI
```

## Anos com duas regras fiscais


```python
# Exibindo os anos correspondentes a essa condio (reutilizando a varivel)

df_1rule = df_brasil[df_brasil['quantity_of_rules_in_place'] == 2]
df_1rule
```

```python
# Calculando o primeiro sub-ndice

multiyear= df_1rule['my_expc_agreggate'] + df_1rule['my_expc_ministries'] + df_1rule['my_expc_line-items']
# Normalizando a varivel:
multiyear = multiyear/3

independent_body = df_1rule['ind_body_sets_budg_assump'] + df_1rule['ind_body_monit_implement']
#Normalizando a variavel:
independent_body = independent_body/2

fiscal_responsability_law = df_1rule['fiscal_responsability_law']

# Construindo o sub-ndice j normalizado para [0,1]
supporting_procedures = (fiscal_responsability_law + independent_body + multiyear)/3 

# supporting_procedures = supporting_procedures.reset_index() isso transforma a series em um dataframe
supporting_procedures
```

```python
# Calculando o segundo sub-ndice normalizado )normalizando para [0,1] dado que temos 2 regras
enforc_procedures = (df_1rule['enforc_nat_DR'] + df_1rule['enforc_nat_ER'] + df_1rule['enforc_nat_BBR'] + df_1rule['enforc_nat_RR'])/2
enforc_procedures

```

```python
# Calculando o terceiro sub-ndice normalizando para [0,1] dado que temos 2 regras

coverage = (df_1rule['cov_nat_DR']/2.5 + df_1rule['cov_nat_ER']/2.5 + df_1rule['cov_nat_BBR']/2.5 + df_1rule['cov_nat_RR']/2.5)/2
coverage
```

```python
# Calculadno o quarto sub-ndice j normalizando para [0,1] dado que temos 2 regras
legal_basis = (df_1rule['lb_nat_DR']/5 + df_1rule['lb_nat_ER']/5 + df_1rule['lb_nat_BBR']/5 + df_1rule['lb_nat_RR']/5)/2
legal_basis
```

```python
# Normalizando os quatro sub-ndices para o intervalo [0,5] e os somando
FRI_2 = (supporting_procedures*5 + enforc_procedures*5 + coverage*5 + legal_basis*5)

# Normalizando o FRI para o intervlao [0,5]
FRI_2=  FRI_2*5/20
FRI_2 = pd.DataFrame(FRI_2)
FRI_2.set_index(df_1rule['Year'], inplace = True)
FRI_2.rename({0: 'FRI'}, axis = 1, inplace = True)
```

```python
df_merged = pd.concat([FRI, FRI_2], ignore_index=True, sort=False)
i = pd.Series([i for i in range(1998, 2022)])
df_merged.set_index(i, inplace = True)
df_merged
```

```python

plt.plot(df_merged, '-')
plt.ylabel('FRI Brasileiro')
plt.xlabel('Ano')
plt.show()
```
