```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Lendo o dataframe preparado somente com os dados do Brasil

```python
# Lendo o dataframe
df_brasil = pd.read_excel('manipulacao_br_python.xlsx')

# Setando o .ipynb para exibir todas as colunas
pd.set_option('display.max_columns', None)

# Filtrando o dataframe para ignorar anos em que no h registro de regras fiscais em vigor
df_brasil = df_brasil[df_brasil['quantity_of_rules_in_place'] != 0]
df_brasil
```

# Trocando todos os valores '-' do df por '0'

```python
df_brasil.replace(to_replace = '-', value = 0, inplace = True)
df_brasil
```

# Calculando o IRF: 4 sub-ndices por ano

## Subndices:

* Supporting procedures
* Formal enforcement procedures
* Coverage
* Legal basis


## Mtodo 2



```python
# Exibindo os anos correspondentes a essa condio

df_1rule = df_brasil[df_brasil['quantity_of_rules_in_place'] == 1]
df_1rule
```

```python
# Calculando o primeiro sub-ndice

multiyear= df_brasil['my_expc_agreggate'] + df_brasil['my_expc_ministries'] + df_brasil['my_expc_line-items']
# Normalizando a varivel:
multiyear = multiyear/3
independent_body = df_brasil['ind_body_sets_budg_assump'] + df_brasil['ind_body_monit_implement']
#Normalizando a variavel:
independent_body = independent_body/2

fiscal_responsability_law = df_brasil['fiscal_responsability_law']

# Construindo o sub-ndice j normalizado para [0,1]
supporting_procedures = (fiscal_responsability_law + independent_body + multiyear)/3 

# supporting_procedures = supporting_procedures.reset_index() isso transforma a series em um dataframe
supporting_procedures

supporting_procedures
```

```python
# Calculando o segundo sub-ndice normalizado 
enforc_procedures = (df_brasil['enforc_nat_DR'] + df_brasil['enforc_nat_ER'] + df_brasil['enforc_nat_BBR'] + df_brasil['enforc_nat_RR'])/df_brasil['quantity_of_rules_in_place']

```

```python
# Calculando o terceiro sub-ndice normalizado
coverage = (df_brasil['cov_nat_DR']/2.5 + df_brasil['cov_nat_ER']/2.5 + df_brasil['cov_nat_BBR']/2.5 + df_brasil['cov_nat_RR']/2.5)/df_brasil['quantity_of_rules_in_place']
coverage
```

```python
# Calculadno o quarto sub-ndice normalizado

legal_basis = (df_brasil['lb_nat_DR']/5 + (df_brasil['lb_nat_ER']-1)/(5-1) + (df_brasil['lb_nat_BBR']-1)/(5-1) + df_brasil['lb_nat_RR']/5)/df_brasil['quantity_of_rules_in_place']
legal_basis

# obs.: na prtica, s estamos normalizando corretamente a legal basis para ER e BBR (pois sabemos, por anlise do dataset, que s h essas duas regras fiscais nesse ano). As outras duas regras fiscais no esto sendo normalizadas pois 0/5 = 0
```

```python
# Plotando a evoluo dos sub-ndices
x = list(range(1998, 2022))

plt.plot(x, supporting_procedures*5, '-')
plt.plot(x, coverage*5, '-')
plt.plot(x, enforc_procedures*5, '-')
plt.plot(x, legal_basis*5, '-')
plt.legend(['supporting procedures', 'coverage', 'enforcement procedures', 'legal basis'])
plt.title('sub-ndices por caracterstica do arcabouo')
plt.ylabel('sub-ndices [0-5] por ano')
plt.xlabel('Ano')
plt.savefig('sub-indices_features.png')
plt.show()
```

```python
# Normalizando os quatro sub-ndices para o intervalo [0,5] e os somando
FRI_met2 = (supporting_procedures*5 + enforc_procedures*5 + coverage*5 + legal_basis*5)

# Normalizando o FRI para o intervlao [0,5]
FRI_met2 = FRI*5/20
FRI_met2 = pd.DataFrame(FRI)

# Transformando a series em um dataframe para colocarmos os outros resultados depois
FRI_met2.set_index(df_brasil['Year'], inplace = True)
FRI_met2.rename({0: 'FRI'}, axis = 1, inplace = True)
FRI_met2.to_excel('dados_FRI_BR.xlsx')
FRI_met2
```

```python

plt.plot(FRI_met2, '-')
plt.title('IRF para o Brasil entre 1998 e 2021: mtodo 2')
plt.ylabel('IRF')
plt.xlabel('Ano')
plt.savefig('FRI_brazil_98-21_met2.png')
plt.show()
```
